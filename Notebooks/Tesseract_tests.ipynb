```python
import pytesseract
import cv2
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = Path("C:\\Users\\yanek\\Dropbox\\PC (2)\\Documents\\20230210_005707.jpg")
image = cv2.cvtColor(cv2.imread(image_path.as_posix()), cv2.COLOR_BGR2RGB)
```

```python
edges = cv2.Canny(image,100,200)
contours, hierarchy =  cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(edges, contours, -1, (0,255,75), 2)
```
