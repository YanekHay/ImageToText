```python
import numpy as np
import torch
from time import time
import random
import matplotlib.pyplot as plt
```

```python
nptimes= []
pytimes= []
tttimes= []

arr = [random.randint(1,500) for i in range(1000000)]
for i in range(100):
    st = time()
    arr = [arr[i]*2 for i in range(len(arr))]
    pytimes.append(time()-st)
    
    arr =  np.array(arr)
    st = time()
    arr*=2
    nptimes.append(time()-st)

    arrt_ = torch.from_numpy(np.array(arr)).cuda()

    st = time()
    arrt_=arrt_*2
    tttimes.append(time()-st)
```

```python
plt.title(' ')
plt.ylabel('()')
plt.xlabel('     ,\n   ')
plt.plot(pytimes,  label=f"Python  {(np.mean(pytimes)*1000):.2f} ")
plt.plot(nptimes, label=f"Numpy  {(np.mean(nptimes)*1000):.2f} ")
plt.plot(tttimes, label=f"Torch  {(np.mean(tttimes)*1000):.2f} ")
plt.legend()
plt.show()
```
