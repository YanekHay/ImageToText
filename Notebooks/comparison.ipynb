```python
import numpy as np
import torch
from time import time
import random
import matplotlib.pyplot as plt
```

```python
nptimes= []
pytimes= []
tttimes= []

arr = [random.randint(1,500) for i in range(1000000)]
for i in range(100):
    st = time()
    arr = [arr[i]*2 for i in range(len(arr))]
    pytimes.append(time()-st)
    
    arr =  np.array(arr)
    st = time()
    arr*=2
    nptimes.append(time()-st)

    arrt_ = torch.from_numpy(np.array(arr)).cuda()

    st = time()
    arrt_=arrt_*2
    tttimes.append(time()-st)
```

```python
plt.title('Բազմապատկման ժամանակը')
plt.ylabel('Ժամանակը(մվ)')
plt.xlabel('Մի քանի անգամ բազմապատկումը կատարվել է,\nորպեսզի երևա ճիշտ ստատիստիկա')
plt.plot(pytimes,  label=f"Python միջինում {(np.mean(pytimes)*1000):.2f} մվ")
plt.plot(nptimes, label=f"Numpy միջինում {(np.mean(nptimes)*1000):.2f} մվ")
plt.plot(tttimes, label=f"Torch միջինում {(np.mean(tttimes)*1000):.2f} մվ")
plt.legend()
plt.show()
```
