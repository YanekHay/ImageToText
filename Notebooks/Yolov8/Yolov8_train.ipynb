# Word/Table Detector

```python
from ultralytics import YOLO

# Load a model
# weight = "yolov8s-seg.pt" # load a pretrained model (SEGMENTATION)
weight = "yolov8s.pt" # load a pretrained model (BBOX)

# data = "./Custom_coco128-seg.yaml" # (SEGMENTATION)
data = "./Custom_box.yaml" # (BBOX)


model = YOLO(weight) # load a pretrained model (recommended for training)
# Train the model
model.train(data=data,
            epochs=20,
            imgsz=1280,
            batch=2,
            save_period=3,
            lr0=0.01,
            lrf=0.003,
            hsv_h=0.015, 
            hsv_s=0.7,
            hsv_v=0.4, 
            translate=0.1, 
            fliplr=0.01,
            mosaic=1.0,
            mixup=0.01,
            optimizer="SGD",
            max_det=1000)
```

# Letter Detector

```python
from ultralytics import YOLO

# Load a model
weight = "yolov8s.pt" # load a pretrained model (recommended for training)
# weight = "/content/drive/MyDrive/Colab Notebooks/Yolo/yolov8s-seg.pt"
# data = "./Custom_coco128-seg_letters.yaml"
data = "./custom_letter_det.yaml"

model = YOLO(weight) # load a pretrained model (recommended for training)
# Train the model
model.train(data=data,
            epochs=50,
            imgsz=320,
            batch=64,
            save_period=2,
            lr0=0.01,
            lrf=0.003,
            hsv_h=0.015, 
            hsv_s=0.7,
            hsv_v=0.4, 
            translate=0.1, 
            fliplr=0.0,
            mosaic=0,
            mixup=0.01,
            optimizer="SGD")
```
