```python
import random
from reportlab.lib import colors
from reportlab.lib.pagesizes import letter, landscape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle
from reportlab.lib.units import inch
from vocabulary.vocabulary import Vocabulary as vb

def generate_random_table():
    # Set up PDF
    doc = SimpleDocTemplate(
        'random_table.pdf',
        pagesize=random.choice([letter, landscape(letter)]),
        rightMargin=72,
        leftMargin=72,
        topMargin=72,
        bottomMargin=18
    )

    # Set up table
    num_rows = random.randint(3, 10)
    num_cols = random.randint(3, 10)
    col_widths = [random.randint(50, 100) for _ in range(num_cols)]
    row_height = max(20, min(col_widths))

    # Generate table data
    data = []
    for _ in range(num_rows):
        row = []
        for _ in range(num_cols):
            value_type = random.choice(['word', 'number'])
            if value_type == 'word':
                language = random.choice(['english', 'armenian'])
                word = vb.random_word(language=language).lower()
                if random.random() < 0.5:
                    word = word.capitalize()
                row.append(word)
            else:
                row.append(str(random.randint(0, 100)))
        data.append(row)

    # Draw table
    table = Table(data, colWidths=col_widths, rowHeights=row_height)
    table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, -1), colors.lightgrey),
        ('TEXTCOLOR', (0, 0), (-1, -1), colors.darkgrey),
        ('FONTNAME', (0, 0), (-1, -1), 'Helvetica'),
        ('FONTSIZE', (0, 0), (-1, -1), 14),
        ('INNERGRID', (0, 0), (-1, -1), 0.25, colors.black),
        ('BOX', (0, 0), (-1,-1), 0.25, colors.black),
    ]))

    # Save PDF
    doc.build([table])

generate_random_table()
```

```python
import random
from reportlab.lib import colors
from reportlab.lib.pagesizes import letter, landscape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph
from reportlab.lib.units import inch
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.lib.styles import getSampleStyleSheet

def generate_random_table():
    # Set up PDF
    doc = SimpleDocTemplate(
        'random_table.pdf',
        pagesize=random.choice([letter, landscape(letter)]),
        rightMargin=72,
        leftMargin=72,
        topMargin=72,
        bottomMargin=18
    )

    # Register DejaVuSans font
    pdfmetrics.registerFont(TTFont('DejaVuSans', 'C:/Users/yanek/AppData/Local/Microsoft/Windows/Fonts/DejaVuSans.ttf'))

    # Set up table
    num_rows = random.randint(3, 10)
    num_cols = random.randint(3, 10)
    col_widths = [doc.width / num_cols] * num_cols

    # Pre-defined lists of random words from Armenian and English languages
    armenian_words = ['լուսին', 'գիրք', 'աղյուս', 'հայր', 'մայր']
    english_words = ['light', 'book', 'salt', 'father', 'mother']
    math_symbols = ['+', '-', '*', '/', '.', '[', ']', '(', ')', '^', '%', '$']

    # Generate table data
    data = []
    row_heights = []
    for _ in range(num_rows):
        row = []
        row_height = 0
        for _ in range(num_cols):
            value_type = random.choice(['word', 'number'])
            if value_type == 'word':
                language = random.choice(['armenian', 'english'])
                if language == 'armenian':
                    words = [random.choice(armenian_words) for _ in range(random.randint(1, 3))]
                else:
                    words = [random.choice(english_words) for _ in range(random.randint(1, 3))]
                if random.random() < 0.5:
                    words = [word.capitalize() for word in words]
                row.append(' '.join(words))
            else:
                number = str(random.randint(0, 100))
                symbol = random.choice(math_symbols)
                row.append(f'{number} {symbol}')
            p = Paragraph(row[-1], getSampleStyleSheet()['Normal'])
            w, h = p.wrap(col_widths[-1], doc.height)
            row_height = max(row_height, h)
        data.append(row)
        row_heights.append(row_height)

    # Draw table
    table = Table(data, colWidths=col_widths, rowHeights=row_heights)
    table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, -1), colors.Color(random.uniform(0.8, 1), random.uniform(0.8, 1), random.uniform(0.8, 1))),
        ('TEXTCOLOR', (0, 0), (-1, -1), colors.Color(random.uniform(0, 0.2), random.uniform(0, 0.2), random.uniform(0, 0.2))),
        ('FONTNAME', (0, 0), (-1, -1), 'DejaVuSans'),
        ('FONTSIZE', (0, 0), (-1, -1), 14),
        ('INNERGRID', (0, 0), (-1, -1), 0.25, colors.black),
        ('BOX', (0, 0), (-1,-1), 0.25, colors.black),
        ('VALIGN', (0, 0), (-1,-1), 'MIDDLE'),
    ]))

    # Save PDF
    doc.build([table])

generate_random_table()
```

```python
def get_characters(cells):
    char_list = []
    for img in cells:
        # Convert to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Threshold the image to create a binary image
        _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

        # Find the contours
        contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Iterate through the contours and extract the characters
        characters = []
        for c in contours:
            x, y, w, h = cv2.boundingRect(c)
            char = img[y:y+h, x:x+w]
            characters.append(char)
        
        char_list.append(characters)
    
    return char_list
```
